```python
import seaborn as sns
import matplotlib.pyplot as plt
roman_data= pd.read_csv("roman-emperor-reigns.csv")
```

```python
roman_data.head()
```

```python
roman_data['Cause_of_Death'].unique()
```

```python
assasinated_df= roman_data[roman_data['Cause_of_Death'].apply(lambda x: 'assassinated' in x.lower())]
```

```python
assasinated_df['Cause_of_Death'].unique()
```

```python
assassinated_no = len(assasinated_df)
other_deaths_no = len(roman_data) - assassinated_no
```

```python
colors = sns.color_palette('pastel')[0:2]
data = [assassinated_no,other_deaths_no]
labels =['Assasinated','Other Deaths']
```

```python
plt.pie(data, labels = labels, colors = colors, autopct='%.0f%%')
plt.legend(loc='upper left')
plt.show()
```
